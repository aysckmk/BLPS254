TASK-5C: Please run the GaussianNB algorithm for 20 different values with respect to varr_smoothing parameter to find the best model. Then please visualize your findings via a lineplot (matplotlib) to illustrate the more appropriate values/ranges.

```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

digits = load_digits()
digits.images.shape

```

```python
fig, axes = plt.subplots(10, 10, figsize=(8, 8),
                         subplot_kw={'xticks':[], 'yticks':[]},
                         gridspec_kw=dict(hspace=0.1, wspace=0.2))

for i, ax in enumerate(axes.flat):
    ax.imshow(digits.images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(digits.target[i]),
            transform=ax.transAxes, color='green')
```

```python
A = digits.data
A.shape
```

```python
B = digits.target
B.shape
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(A, B, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB                           
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import seaborn as sns
from numpy import random

iris = sns.load_dataset('iris')
X_iris= iris.drop('species', axis=1)
y_iris = iris['species']

Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=0)
```

```python
random.random
for i in range(20):
    print("{:.6f}".format(random.random()))
```

```python
Values=[0.787243,
0.641518,
0.914810,
0.058406,
0.843949,
0.504119,
0.221992,
0.251530,
0.061851,
0.589527,
0.764826,
0.366688,
0.787932,
0.290760,
0.872502,
0.859003,
0.867360,
0.267302,
0.723131,
0.148106]

for i in range (20):
  V = Values[i]
  model = GaussianNB(var_smoothing=V)                       
  model.fit(Xtrain, ytrain)                  
  y_model = model.predict(Xtest)
  acc_scr = (accuracy_score(ytest, y_model)) 
  print(acc_scr)
  

plt.plot(Values)
plt.show()
```

```python
print("The best accuracy score: ", max(Values))
```
