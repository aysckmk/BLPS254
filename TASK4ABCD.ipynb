4B)Please provide the visual illustrations of the dataset that will demonstrate the insightful distributions with respect to the target column (CDRGLOB: the level of the Alzheimer Disease -AD)

```python
x = data["CERECSF"]
y = data["TEMPCOR"]
labels = data["CDRGLOB"]
plt.scatter(x,y, cmap='viridis',c=labels, alpha=0.5)
plt.xlabel('CERECSF Level')
plt.ylabel('TEMPCOR Index')
plt.title('TEMPCOR-CERECSF Scatter')
plt.colorbar()
plt.show()
```

```python
x = data["HIPPOVOL"]
y = data["INDEPEND"]
labels = data["CDRGLOB"]
plt.scatter(x,y, cmap='viridis',c=labels, alpha=0.5)
plt.xlabel('Hippovol Level')
plt.ylabel('Independ Index')
plt.title('HIPPOVOL-INDEPEND Scatter')
plt.colorbar()
plt.show()
```
