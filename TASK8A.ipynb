```python
# -*- coding: utf-8 -*-

import time

class Pclab():
    def __init__(self, lab_name="Mobile Lab", lab_status="Açık", lab_students=["2221","2223","2228","2120","2102","2229","2187","22157","2207","2245","2134","2203","2109"], numberof_students=13, numberof_freepc=7, numberof_busypc=13):
        self.lab_name = lab_name
        self.lab_status=lab_status
        self.lab_students = lab_students
        self.numberof_students = numberof_students
        self.numberof_freepc = numberof_freepc
        self.numberof_busypc= numberof_busypc

    def info(self):
        print("Lab Name: {}\n Lab Status: {}\n Lab Students: {}\nNumber of Students: {}\nNumber of Free Pc: {}\nNumber of Busy Pc: {}".format(self.lab_name,self.lab_status,self.lab_students,self.numberof_students,self.numberof_freepc,self.numberof_busypc))




class Asisstant(Pclab):
    def lab_open(self):
        if (self.lab_status == "Close"):
            self.lab_status = "Kapalı"
            print("Lab Opening...")
        else:
            print("Lab is Open!")
    
    def lab_close(self):
        if (self.lab_durum == "Open"):
            self.lab_durum = "Close"
            print("Lab closing...")
        else:
            print("Lab  closed!")
    
    def add_student(self, std_no):
        if std_no in self.lab_students:
            print("This student is already on the list")
        else:
            print("Adding students to the list")
            time.sleep(1)
            self.lab_students.append(std_no)
            self.numberof_students += 1
            print("Students added.")
            print("List of Student: {}".format(self.lab_students))

    def listof_student(self):
        print(self.lab_students)



class Students(Pclab):
    def pc_reserve(self):
        if (self.numberof_busypc> 0):
            print("Computer is being reserved...")
            time.sleep(1)
            self.numberof_freepc -= 1
            self.numberof_busypc += 1
        else:
            print("There are no free computers in the lab.")

    def pc_clear(self):
        print("Leaving the computer")
        time.sleep(1)
        self.numberof_freepc -= 1
        self.numberof_busypc += 1
        print("Ther are free computer in the lab.")


lab=Pclab()
Students=Students(Pclab)
asisstant=Asisstant(Pclab)

while True:
    selection = input("Please select the person.\n 1-Assistant\n\n\n\n2-Student \n")
    if (selection == "1"):

        print("""
            1.Add Student
            2.List of Student
            3.Informations
            4.Open Lab
            5.Closed Lab
            6.Exit
            
            """)

        process1 = input("Welcome, please make a selection for your transaction")

        if process1 == "1":
            std_no = input("Enter student number.")
            asisstant.add_student(std_no)
        elif process1 == "2":
            asisstant.listof_student()
        elif process1 == "3":
            lab.info()
        elif process1 == "4":
            asisstant.lab_open()
            print("Lab is open!")
        elif process1 == "5":
            asisstant.lab_close()
            print("Lab is closed.")
        elif process1== "6":
          break
        else:
            print("You selected an invalid transaction.")

    elif (selection == "2"):
        print("""
        1.Sit at a computer desk.
        2.Get up from a computer desk.
        3.Show information
        Exit 4.
        """)
        process2 = input("Welcome, please make a selection for your transaction")
        if process2 == "1":
            Students.pc_reserve()
        elif process2 == "2":
            Students.info()
        elif process2 == "4":
            break
        else:
            print("You selected an invalid transaction.")
```

```python
import time
import random



class Kumanda():

    def __init__(self, tv_durum="Kapalı",tv_ses=0,kanal_listesi=["TRT"],kanal="TRT",ınternet="Bağlı Değil!",youtube="Kapalı",yazılım_guncelleme= "Kapalı"):

        self.tv_durum=tv_durum
        self.tv_ses=tv_ses
        self.kanal_listesi=kanal_listesi
        self.kanal= kanal
        self.ınternet=ınternet
        self.youtube=youtube
        self.yazılım_guncelle =yazılım_guncelleme

    def tv_ac(self):

        if (self.tv_durum  == "Kapalı"):
            print("Tv Açılıyor...")
            self.tv_durum="Açık"
        else:
            print("Tv Açık")

    def tv_kapat(self):
        
        if (self.tv_durum  == "Kapalı"):
            print("Tv Kapalı")
        
        else:
            print("Tv Kapatılıyor...")
            self.tv_durum="Kapalı"
    

    def ses_ayari(self):

        while True:

            cevap=input("Sesi artır :>\nSesi azalt: <\nÇıkış: q")

            if (cevap=='>'):
                if self.tv_ses>=100:
                    print("Max Ses")

                self.tv_ses+=1
                print("Tv Sesi: {}".format(self.tv_ses))
            
            elif (cevap=="<"):
                if (self.tv_ses<=0):
                    print("Min ses")

                self.tv_ses-=1
                print("tv_ses: {}",format(self.tv_ses))
            elif (cevap=="q"):
                break
            else:
                print("Hatalı Tuşlama")

    def kanal_ekle(self,kanal):
        print("Kanal ekleniyor...")
        time.sleep(1)

        self.kanal_listesi.append(kanal)

        print("Kanal listesi: {}".format(self.kanal_listesi))

    def kanal_sec(self):

        rastgele=random.randint(0,len(self.kanal_listesi)-1)

        self.kanal=self.kanal_listesi[rastgele]
        print(self.kanal)
    
    def __len__(self):
        return len(self.kanal_listesi)

    def Connect_Youtube(self):
        while True:
            if self.youtube == "Açık":
                print("Youtube Açık!")
                break
            elif self.youtube == "Kapalı":
                cevap = input("Youtube'a bağlanmak istiyor musunuz? (E/H)")
                if cevap == "E" and self.ınternet == "Bağlı Değil!":
                    print("İnternet Bağlı Değil!")
                    cevap2 = input("İnternete Bağlanmak İstiyor musunuz? (E/H)")
                    if cevap2 == "E":
                        print("İnternete Bağlanıyor...")
                        time.sleep(3)
                        self.ınternet = "Bağlı!"
                        print("İnternete Bağlandınız!")
                        print("Youtube'a Bağlanılıyor...")
                        time.sleep(1)
                        self.youtube = "Açık"
                        print("Youtube Açıldı!")
                        break
                    elif cevap2 == "H":
                        print("İnternete Bağlanılamadı, Tekrar deneyiniz...")
                        time.sleep(1)
                        break
                    else:
                        print("Lütfen E veya H harflerinden birini giriniz.")
                        continue
                elif cevap == "E" and self.ınternet == "Bağlı!":
                    print("İnternete bağlısınız, Youtube Açılıyor...")
                    self.youtube = "Açık"
                    time.sleep(1)
                    print("Youtube Açıldı!")

                elif cevap == "H":
                    print("Youtube'a bağlanmayı reddettiniz, Özellik kapatılıyor...")
                    time.sleep(1)
                    break

                else:
                    print("Lütfen E veya H harflerinden birini giriniz.")
                    continue

      
    def yazılımı_guncelle(self):
        while True:
            cevap = input("Yazılım güncellemesini kontrol etmek istiyor musunuz? (E/H)")
            if cevap == "E" and  self.ınternet=="Bağlı!":
                if self.yazılım_guncelle == "Açık":
                    print("Yazılımınız güncel.")
                    break

                else:
                    print("Yazılım güncellemesi başlatılıyor...")
                    self.yazılım_guncelle = "Açık"
                    time.sleep(2)
                    print("Yazılım güncellendi")
                    continue
            elif cevap == "E" and self.ınternet == "Bağlı Değil!":
                print("İnternet Bağlı Değilsiniz!")
                cevap2 = input("İnternete Bağlanmak İstiyor musunuz? (E/H)")
                if cevap2 == "E":
                    print("İnternete Bağlanıyor...")
                    time.sleep(3)
                    self.ınternet = "Bağlı!"
                    if self.yazılım_guncelle == "Açık":
                        print("Yazılımınız güncel.")
                        break

                    else:
                        print("Yazılım güncellemesi açılıyor...")
                        self.yazılım_guncelle = "Açık"
                        time.sleep(2)
                        print("Yazılım güncellendi")
                elif cevap2 == "H":
                    print("İnternete Bağlanılmadı, İnternet olmadan Yazılım güncelleme hizmeti verilememektesir! Özellik kapatılıyor...")
                    time.sleep(1)
                    break

                else:
                    print("Lütfen E veya H harflerinden birini giriniz.")
                    continue

            elif cevap == "H":
                print("Yazılım güncellemesini açmayı reddettiniz, kumanda kapatılıyor.")
                break
            else:
                print("Lütfen E veya H harflerinden birini giriniz.")
                continue

    def __str__(self):
        return "Tv_durum: {}\ntv_ses: {}\nkanal listesi: {}\nkanal:{}\ninternet:{}\nyoutube:{}\nyazılım_guncelligi:{}".format(self.tv_durum,self.tv_ses,self.kanal_listesi,self.kanal,self.ınternet,self.youtube,self.yazılım_guncelle)



print("""
        1. TV aç
        2. TV kapat
        3. Ses Ayarları
        4. Kanal Ekle
        5. Açık Kanalı Öğren
        6. Kanal Sayısı
        7. TV Bilgileri
        8.Youtube'a Bağlan
        9.Yazılım Güncellemesi
Çıkmak için q'ya bas.
""")
kumanda=Kumanda()

while True:

    islem= input("İslem seçiniz")

    if islem =="q":
        print("Sonlandırılıyor")
        break 
    elif islem =="1":
        kumanda.tv_ac()
    
    elif islem=="2":
        kumanda.tv_kapat()

    elif islem=="3":
        kumanda.ses_ayari()
    
    elif islem=="4":

        kanal_isimleri=input("Kanal isimlerini lütfen ',' ile ayırarak giriniz.")

        x=kanal_isimleri.split(",")


        for i in x:
            kumanda.kanal_ekle(i)
    
    elif islem=="5":
        kumanda.kanal_sec()

    elif islem=="6":
        print("Kanal sayısı: ",len(kumanda))
    
    elif islem=="7":
        print(kumanda)

    elif islem=="8":
        kumanda.Connect_Youtube()


       
    elif islem == "19":
        kumanda.yazılımı_guncelle()

    else:
        print("hatalı Tuşlama")
```
